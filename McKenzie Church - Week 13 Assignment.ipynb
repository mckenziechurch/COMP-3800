**1. What is the difference between supervised and unsupervised learning? Provide an example for each type.  10 points**

- Supervised learning: method where the dataset is labelled for the algorithm; the prediction relies on the labels to guide the prediction and impacts accuracy
    - ie, logistic regression for dependent target and independent variables

- Unsupervised learning: method where the models are trained with an *unlabelled* dataset; prediction is based on the structure of the dataset rather than categories
    - ie, clustering for determining which data is similar to each other

**3. With the above two clusters you find, calculate the Silhouette Coefficient of Point c. 10 points**

```python
import math
c = [2.1, 1] # comparison point
b = [1, 1]
a = [1.5, 2]
d= [3, 2] # this is the centroid for cluster 2
e = [3.9, 1]
f = [3.1, 2.5]

# cluster 1:
print("Cluster 1:")
cb = math.dist(c, b) # calculate distance btwn c and b
ca = math.dist(c, a) # calculate distance btwn c and a
print("c to b: " + str(cb))
print("c to a: " + str(ca))
avg1 = (cb + ca) / 2 # calculate avg distances in cluster 1
print("Avg: " + str(avg1))
print("\n")

# cluster 2:
print("Cluster 2:")
ce = math.dist(c, e) # calculate distance btwn c and e
cf = math.dist(c, f) # calculate distance btwn c and f
cd = math.dist(c, d) # calculate distance btwn c and d
print("c to e: " + str(ce))
print("c to f: " + str(cf))
print("c to d: " + str(cd))
avg2 = (ce + cf + cd) / 3 # calculate avg distances in cluster 2
print("Avg: " + str(avg2))
```

```python
# cluster 1's average distance is closer to c so we will use this for silh. coef.
sil_coef = 1 - (avg1 / avg2)
print("Silhouette Coef. - point c: " + str(sil_coef))
```

**4. Calculate the SSE of the following two clustering methods in the picture. SSE is the sum of the squared differences between each observation and its group's mean. Which clustering method is better? 10 points**


```python
def calculateSSE(nums, avg):
    diff = []
    # calculate the sum of squared diff btwn each point and the cluster's mean
    for i in nums:
        diff_square = (i - avg)**2
        diff.append(diff_square)
    # calculate the sse using sums
    sse = sum(diff)
    print(sse)
```

**Line 1**:
- has two larger clusters of w/ numbers from range(-5, 6)

```python
# BLUE CLUSTER

# get all observations
line1_blue = [-5, -4, -2, -1]

# get the mean
line1_blue_mean = sum(line1_blue) / len(line1_blue)

print("Line 1, BLUE Cluster Avg: " + str(line1_blue_mean) + "\n")
print("SSE of Line 1, BLUE Cluster: ")
calculateSSE(line1_blue, line1_blue_mean)
```

```python
# ORANGE CLUTER
line1_orange = [2, 3, 4]
line1_orange_mean = sum(line1_orange) / len(line1_orange)

print("Line 1, ORANGE Cluster Avg: " + str(line1_orange_mean) + "\n")
print("SSE of Line 1, ORANGE Cluster: ")
calculateSSE(line1_orange, line1_orange_mean)
```

**Line 2**:
- has three clusters w/ same amount of data as line 1

```python
# BLUE CLUSTER
line2_blue = [-5, -4]
line2_blue_mean = sum(line2_blue) / len(line2_blue)

print("Line 2, BLUE Cluster Avg: " + str(line2_blue_mean) + "\n")
print("SSE of Line 2, BLUE Cluster: ")
calculateSSE(line2_blue, line2_blue_mean)
```

```python
# GREEN CLUSTER
line2_green = [-2, -1]
line2_green_mean = sum(line2_green) / len(line2_green)

print("Line 2, GREEN Cluster Avg: " + str(line2_green_mean) + "\n")
print("SSE of Line 2, GREEN Cluster: ")
calculateSSE(line2_green, line2_green_mean)
```

```python
# ORANGE CLUSTER
line2_orange = [2, 3, 4]
line2_orange_mean = sum(line2_orange) / len(line2_orange)

print("Line 2, ORANGE Cluster Avg: " + str(line2_orange_mean) + "\n")
print("SSE of Line 2, ORANGE Cluster: ")
calculateSSE(line2_orange, line2_orange_mean)
```

The clustering method from line 2 is better because the SSEs calculated from each cluster are smaller than those in line 1, which means that the individual point and the prediction have a smaller margin (distance) between them. 

**5. How feature reduction work differently for a supervised-learning problem and an unsupervised-learning problem? 10 points**

- Supervised learning does feature reduction by looking at the correlation between each invidiual variable and the target variable, then dropping the ones with low correlation to the target variable. This is looking at probability of a specified outcome.

- Unsupervised learning reduces by looking at the *structure* of the data to determine which data are correlated with *each other*, rather than a specific target variable/goal. This is looking at the correlation of data between each other.

**9. Apply the K-means clustering algorithm. Use the elbow method to find the best k. Plot how the Sums of Square Error change as k increases. What is the best k? 10 points**

```python
# build kmeans model
from sklearn.cluster import KMeans
X = np.array(list(zip(new_df['Malic_Acid'], new_df['Alcohol'], new_df['Ash'], new_df['Ash_Alcanity'], 
                     new_df['Color_Intensity'], new_df['Magnesium'], new_df['Total_Phenols'])))
    
kmeans = KMeans(n_clusters=3)
kmeans = kmeans.fit(X)
```

```python
# use model to find the groups for each point and centroids
labels = kmeans.predict(X)
centroids = kmeans.cluster_centers_

for x, y in zip(labels, centroids):
    print(x, y, "\n")
```

```python
# calculate the sse - watch the sse between as k increases in the range
# k = num of clusters
sse = [] 

for k in range(1, 61):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)
for x, y in zip((range(1, 61)), sse):
    print(x, y)
```

```python
# plot a visual graph of relation btwn sse and k - find k for elbow (optimal k)
plt.plot(range(1, 61), sse)
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()

# the elbow appears to be for 0 < k < 10 - let's verify
```

```python
# mathematically find the optimal k (elbow)
import kneed
ks = kneed.KneeLocator(range(1, 61), sse, curve="convex", direction="decreasing")
print("Optimal K: " + str(ks.elbow))
```

**10. Use the Silhouette Coefficient method to find the optimal k. Does it return the same k as the elbow method? 10 points**

```python
from sklearn import metrics
sil = []
for k in range(2, 62):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    score = metrics.silhouette_score(X, kmeans.labels_)
    sil.append(score)
    print(k, " ", score )
```

```python
# plot the silho. coef. for each k 
plt.plot(range(2, 62), sil)
plt.xlabel("Number of Clusters")
plt.ylabel("silhouette_score")
plt.show()
```

The k for the Elbow Method and Silhouette Coeff. are significantly different, with the elbow method containing scores that are higher.
