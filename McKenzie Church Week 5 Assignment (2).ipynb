**1. Read the data from https://raw.githubusercontent.com/prust/wikipedia-movie-data/master/movies.json . Check the top 10 entries to get familiar with the data. Convert the data into a Pandas DataFrame   10 points**

```python
import json
import requests
req = requests.get('https://raw.githubusercontent.com/prust/wikipedia-movie-data/master/movies.json')
movies = json.loads(req.text)
type(movies)
```

      - Then you will be asked to scrape the data from all the pages.**
      
**5. Scrape the html content from the first page   5 points**

```python
from bs4 import BeautifulSoup as bs
url = 'https://www.whitehouse.gov/briefing-room/statements-releases/page/1/'
req = requests.get(url)
soup = bs(req.text, 'html.parser')
print(soup.prettify())
```

**6. Use find_all() to extract all the titles and the dates. Create an empty DataFrame df_briefings.  Save this information to the DataFrame. 10 points**

```python
#titles and dates from page 1
titles = soup.find_all('a', attrs={'class':'news-item__title'})
dates = soup.find_all('time', attrs={'class':'posted-on entry-date published updated'})
df_briefings = pd.DataFrame(columns=['titles', 'dates'])
for title, date in zip(titles, dates):
    df_briefings.loc[len(df_briefings)] = [title.text, date.text]
df_briefings
```

**7.  Reuse the above code. Use a for loop to scrape the titles and dates of all the available statements on that site. Save it to a new DataFrame,  df_briefingsAll.  15 points**

```python
#do all page numbers
df_allpages_briefings = pd.DataFrame(columns=['titles', 'dates'])
url = 'https://www.whitehouse.gov/briefing-room/statements-releases/page/'
soup = bs(req.text, 'html.parser')
for page in range(1,11):
    req = requests.get(url + str(page))
    soup = bs(req.text, 'html.parser')
    titles = soup.find_all('a', attrs={'class':'news-item__title'})
    dates = soup.find_all('time', attrs={'class':'posted-on entry-date published updated'})
    #df_briefings = pd.DataFrame(columns=['titles', 'dates'])
    for title, date in zip(titles, dates):
        df_allpages_briefings.loc[len(df_allpages_briefings)] = [title.text, date.text]
df_allpages_briefings
```

**8. On which date/dates the White House posted the most statements?  10 points**

```python
df_briefingsAll_pivot = pd.pivot_table(df_allpages_briefings, values='titles', index='dates', aggfunc='count', sort=True)
df_briefingsAll_pivot.nlargest(5, 'titles')
```

Date(s) w/ Most Statements: September 02, 2022
