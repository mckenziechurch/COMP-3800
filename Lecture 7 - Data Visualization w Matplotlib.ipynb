```python
#import libraries, matplotlib
#upload downloaded file in jupyter notebook home page
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

df = pd.read_csv('sales.csv')
```

```python
print(df)
```

## Basic Bar Plot

```python
plt.bar(df.iloc[:,0], df.iloc[:,6]) #basic bar plot
```

```python
plt.bar(df.iloc[:,0], df.iloc[:,6])

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')

plt.yticks(np.arange(0, 20, 2))
plt.show
```

## Switched Axis on Basic Bar Plot

```python
plt.barh(df.iloc[:,0], df.iloc[:,6]) #switch graph arrangement

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')

plt.xticks(np.arange(0, 20, 2)) #range, exclusive, step
plt.show()
```

## Grouped Bar Plot

```python
#do records for each person bar chart - grouped bar chart
#plt.bar(x, heigh, width, bottom, alight); x = position on the axis (starting and ending for individual bar)

width = 0.1
x_axis = np.arange(0, 4) #gives us number of groups (number of positions)

plt.bar(x_axis, df.iloc[:,1], width = width) #position for first group/person

plt.bar(x_axis+0.1, df.iloc[:,2], width  = width) #position for second group/person...etc

plt.bar(x_axis+0.1*2, df.iloc[:,3], width  = width) #multiply to separate group positions

plt.bar(x_axis+0.1*3, df.iloc[:,4], width  = width)

plt.bar(x_axis+0.1*4, df.iloc[:,5], width  = width) #each for a day of the week in this example
plt.plot(df.iloc[:,6]) #adds a line above the bar chart; must change legend also

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')

plt.xticks(x_axis, df.iloc[:,0]) #give names of each group for the groups
plt.yticks(np.arange(0, 10, 1)) 

plt.legend(df.columns[1:6], loc = 'best')
plt.show()
```

## Stacked Bar Plot

```python
x_axis = np.arange(4)
width = 0.15

plt.bar(x_axis, df.iloc[:, 1], width = width) 
plt.bar(x_axis, df.iloc[:, 2], width = width,  bottom = df.iloc[:, 1])  
# For stacked bar plot, the bottom of the bars are raised; x doesn't change, the bottom of the bar positions will change

plt.bar(x_axis, df.iloc[:, 3], width = width, bottom = df.iloc[:, 1]+df.iloc[:, 2])
plt.bar(x_axis, df.iloc[:, 4], width = width, bottom = df.iloc[:, 1]+df.iloc[:, 2]+df.iloc[:, 3])
plt.bar(x_axis, df.iloc[:, 5], width = width, bottom = df.iloc[:, 1]+df.iloc[:, 2]+df.iloc[:, 3]+df.iloc[:, 4])

plt.xlabel('Sales Representatives')
plt.ylabel('No# of Products')
plt.title('Sales Records')

plt.xticks(x_axis, df.iloc[:,0]) 
plt.yticks(np.arange(0, 10, 1)) 

plt.legend(df.columns[1:6], loc = 'best')

plt.show()
```

## Pie Chart

```python
#all are contained within the plt.pie method with commas between
plt.pie(df.iloc[:, 6],   # data source
        explode = [0.05, 0.05, 0.2, 0.05],  # this separates the pies from the center
        labels = df.iloc[:, 0],  # labels on the pie like Anna, Jason
        shadow = True, 
        autopct = '%.2f%%')  # the percentage numbers on the pie
plt.legend()
plt.show()
```

```python
#without explode
plt.pie(df.iloc[:, 6],   # data source
        labels = df.iloc[:, 0],
        shadow = True, 
        autopct = '%.2f%%')
plt.legend()
plt.show()
```

```python
plt.pie(df.iloc[:, 6],   # data source
        explode = [0, 0, 0.4, 0],  # explode only one data in the pie chart
        labels = df.iloc[:, 0],
        shadow = True, 
        autopct = '%.2f%%') 
plt.legend()
plt.show()
```

## Adding Subplots

```python
#can use nested for loop to shorten code
fig, axes = plt.subplots(2,3, sharex = True, sharey = True)
  # This creates a 2*3 canvas for subplots

axes[0, 0].bar(x_axis, df.iloc[:, 2], width = width,  bottom = df.iloc[:, 1])
axes[0, 0].set_title('Monday')

axes[0, 1].bar(x_axis, df.iloc[:, 3], width = width, bottom = df.iloc[:, 1]+df.iloc[:, 2])
axes[0, 1].set_title('Tuesday')

axes[0, 2].bar(x_axis, df.iloc[:, 4], width = width, bottom = df.iloc[:, 1]+df.iloc[:, 2]+df.iloc[:, 3])
axes[0, 2].set_title('Wednesday')

axes[1, 0].bar(x_axis, df.iloc[:, 5], width = width, bottom = df.iloc[:, 1]+df.iloc[:, 2]+df.iloc[:, 3]+df.iloc[:, 4])
axes[1, 0].set_title('Thursday')

fig.suptitle('Sales Records')

plt.show()
plt.subplots_adjust(wspace=0, hspace=0)
```
